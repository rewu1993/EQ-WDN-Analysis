# Map pipe failure rates to the network for each scenario

```python
from scipy.spatial import cKDTree
from shapely.geometry import Point, LineString
import itertools
from operator import itemgetter

def ckdnearest(gdfA, gdfB, gdfB_cols=['pgv']):
    A = np.concatenate(
        [np.array(geom.coords) for geom in gdfA.geometry.to_list()])
    B = [np.array(geom.coords) for geom in gdfB.geometry.to_list()]
    B_ix = tuple(itertools.chain.from_iterable(
        [itertools.repeat(i, x) for i, x in enumerate(list(map(len, B)))]))
    B = np.concatenate(B)
    ckd_tree = cKDTree(B)
    dist, idx = ckd_tree.query(A, k=1)
    idx = itemgetter(*idx)(B_ix)
    gdf = pd.concat(
        [gdfA, gdfB.loc[idx, gdfB_cols].reset_index(drop=True)], axis=1)
    return gdf

def pgv_node2pipe(pipe_info,node_info):
    pgvs = []
    for index, row in pipe_info.iterrows():
        nodes = [str(row['node1']),str(row['node2'])]
        end_nodes_info = node_info.loc[node_info['node_id'].isin(nodes)]
        pgv = np.mean(end_nodes_info['pgv'])
        pgvs.append(pgv)
    return pgvs


def get_prefix(file_path):
    file_name = file_path.split('/')[-1]
    prefix = file_name.split('.')[0]
    return prefix

def add_pgv2pipe(pgv_path):
    pgv_info = pd.read_csv(pgv_path)
    gd_pgv = gpd.GeoDataFrame(
        pgv_info, geometry=gpd.points_from_xy(pgv_info.lon, pgv_info.lat))
    df = ckdnearest(node_info,gd_pgv)
    pgvs = pgv_node2pipe(pipe_info,df)
    pipe_info['pgv'] = pgvs
    return pipe_info
    

    



    
```

```python
k_dict = {'A':1,
          'C': 1, 
          'D':0.5, 
          'F':1, 
          'H':1, 
          'K':1, 
          'N':1, 
          None:1,
          'S':0.6,
          'W':1}
def cm2inch(cm):
    return 39.3701*cm/100

def calculate_fail_prob(k,pgv,l):
    rr = k*0.00187*pgv/1000
    failure_rate = 1- np.power(np.e,-rr*l)
    return failure_rate

def get_pipe_failrate(pipe_info):
    fail_probs = []
    for index, row in pipe_info.iterrows():
        m,l,pgv = row['material'],row['length'],row['pgv']
        prob = calculate_fail_prob(k_dict[m],l,cm2inch(pgv))
        fail_probs.append(prob)
    return fail_probs

def add_failrate2pipe(pgv_path):
    pipe_info = add_pgv2pipe(pgv_path)
    pipe_info['fail_prob'] = get_pipe_failrate(pipe_info)
    
    
    pgv_prefix = get_prefix(pgv_path)
    save_path = save_folder + 'pipes_'+ pgv_prefix +'.geojson'
    pipe_info.to_file(save_path, driver="GeoJSON")
    print (f'saved to {save_path}')
```

# Analyze PGA/PGV data 

```python
def analyz_im(im,bound):
    im.sort()
    top_im = im[bound:]
    mean = np.mean(top_im)
    std = np.std(top_im)
    return mean

    
```

```python
case_names = []
pga_res = []
pgv_res = []
for pgv_csv_file in pgv_csv_files:
    df = pd.read_csv(pgv_csv_file)
    case_name = get_prefix(pgv_csv_file)
    case_names.append(case_name)
    print (case_name)
    
    pga = df['pga'].to_numpy()
    pgv = df['pgv'].to_numpy()
    bound = int(len(pga)*0.)
    
    pga_res.append(analyz_im(pga,bound))
    pgv_res.append(analyz_im(pgv,bound))
    
    
```

```python
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 12})
plt.subplot(1, 2, 1)
plt.hist(pgv_res)
plt.xlabel('PGV (cm/s)')
plt.ylabel('Count')

plt.subplot(1, 2, 2)
plt.hist(pga_res)
plt.xlabel('PGA (g)')
# plt.ylabel('Count')
plt.tight_layout()
plt.savefig('../data/rupture/figures/ims_overall.png',bbox_inches='tight',dpi=300)
plt.show()
```
