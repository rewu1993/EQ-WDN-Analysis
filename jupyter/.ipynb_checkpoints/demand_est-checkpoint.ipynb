```python
demand_dict = {'residential': 1500/(24*60),
              'commercial':4000/(24*60)}
NODE_REG = {}
```

```python
def m2_acre(a_m2):
    return a_m2/4046.8

def calc_demand(land_type,area):
    return area*demand_dict[land_type]
```

```python
def find_discharge_node(block_id,block_node_info):
    block = block_node_info[block_node_info['block_id'] == block_id]
    return block['node_id'].item()
```

```python
alameda_block_info['type'] = len(alameda_block_info)*['residential']

min_area = 0.1
max_area = 20
num_valid_area = 0
node_demand = {}
for i, row in alameda_block_info.iterrows():
    area_m2 = row['area']
    land_type = row['type']
    block_id = row['block_id']
    area_acre = m2_acre(area_m2)
    
    if ((area_acre> min_area) and 
        (area_acre< max_area)):
        num_valid_area+=1
        demand = calc_demand(land_type,area_acre)
        node_id = find_discharge_node(block_id,block_node_info)
        if node_id not in node_demand:
            node_demand[node_id] = demand
        else: 
            node_demand[node_id] += demand
print (num_valid_area,len(alameda_block_info)-num_valid_area)
```
